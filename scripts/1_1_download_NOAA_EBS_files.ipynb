In this script, data on Hematodinium infection from the 2014-2019 NOAA Eastern Bering Sea trawl survey index stations are downloaded and examined

First, download data from Gannett

```python
!wget --no-check-certificate --no-parent --recursive  --reject "index.html" -nd https://gannet.fish.washington.edu/hematodinium/Hemato%20samples/data%202014-2019/ -P ../data/NOAA_EBS_trawl_survey/
```

Remove all files that aren't the data files we want

```python
!find ../data/NOAA_EBS_trawl_survey/ -type f ! -name "201*BCS Index Station Collection Data_MASTER*" -delete
```

Replace all spaces in filenames with underscores

```python
! for file in ../data/NOAA_EBS_trawl_survey/*; do mv "$file" `echo $file | tr ' ' '_'` ; done
```

Check all filenames and get size

```python
ls -sh ../data/NOAA_EBS_trawl_survey/*
```
